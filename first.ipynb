```python
premium_customer = pd.DataFrame(data = {'Columns_Name': ['Mainstream', 'Budget', 'Premium'], 
                                        'Value_Count': [29245,24470,18922]})
premium_customer
```

```python
import seaborn as sns
def addlabels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i])
plt.style.use('ggplot')
sns.barplot(data = premium_customer,x='Columns_Name', y='Value_Count')
addlabels(premium_customer['Columns_Name'], premium_customer['Value_Count'])
plt.xlabel('Column_Names')
plt.ylabel('Value_Count')
plt.title('Premium Customer Category Numbers')
```

```python
life_st = pd.DataFrame(data = {'Columns_name': ['Retires', 'Older S/C', 'Young S/C', 'Older F', 'Young F',
                                                'Midage S/C', 'New F'], 'Value_Count': [14805, 14609, 14441, 9780, 9178, 7275, 2549]})
life_st
```
